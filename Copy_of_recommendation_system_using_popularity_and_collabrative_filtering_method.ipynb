```python
#collaborative filtering method
```

```python
rating_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_user=x[x].index
padhe_likhe_user
```

```python
rating_with_name[rating_with_name['User-ID']==8]
```

```python
filtered_data=rating_with_name[rating_with_name['User-ID'].isin(padhe_likhe_user)]
filtered_data
```

```python
y=filtered_data.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
famous_books
```

```python
final_rating=filtered_data[filtered_data['Book-Title'].isin(famous_books)]
final_rating
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
import numpy as np
```

```python
def recommend(book_name):
  index=np.where(pt.index==book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
  for i in similar_items:
    print(pt.index[i[0]])
recommend('1984')
```
